```python
import os 
import pandas as pd
```

```python
class Return_invisible:
    def __init__(self,path,n=int,**kwargs):
        if os.path.exists(path)==False:
            raise Exception('no data in the path.')
        self.df=pd.read_csv(path)      
        
        self.df['conversion']=(self.df['WebVisit']+
                               self.df['CollateralView']+
                               self.df['ProductView']+
                               self.df['FormComplete'])
        self.advance=kwargs['advance']
        if self.advance:
            self.df1=self.df[['Date',
                             'AudienceType',
                             'AdDigitalChannel',
                             'VideoCompletes',
                             'SocialLikes',
                             'SocialShares',
                             'conversion']]
            self.df2=self.df[['Date',
                             'AudienceType',
                             'AdDigitalChannel',
                             'CTR']]

            self.df1=self.df1.groupby(['Date','AudienceType','AdDigitalChannel']).sum()
            self.df1=self.df1.reset_index()
            self.df2=self.df2.groupby(['Date','AudienceType','AdDigitalChannel']).mean()
            self.df2=self.df2.reset_index()
            self.df=pd.concat([self.df1,self.df2['CTR']],axis=1)
            self.df=self.df.loc[:,['Date',
                                   'AudienceType',
                                   'AdDigitalChannel',
                                   'VideoCompletes',
                                   'SocialLikes',
                                   'SocialShares',
                                   'CTR',
                                   'conversion']]

        else:
            self.df1=self.df[['Date',
                             'VideoCompletes',
                             'SocialLikes',
                             'SocialShares',
                             'conversion']]
            self.df2=self.df[['Date',
                             'CTR']]

            self.df1=self.df1.groupby(['Date']).sum()
            self.df1=self.df1.reset_index()
            self.df2=self.df2.groupby(['Date']).mean()
            self.df2=self.df2.reset_index()
            
            self.df=pd.concat([self.df1,self.df2['CTR']],axis=1)

            self.df=self.df.loc[:,['Date',
                                   'VideoCompletes',
                                   'SocialLikes',
                                   'SocialShares',
                                   'CTR',
                                   'conversion']]
        self.n=n
        
        
    def __call__(self):
        print('save the return 2 in the Output folder')
        if self.advance==False:
            self.output=self.ndays(self.df,self.n)
            self.output.to_csv('Output/return2_'+str(self.n)+'days.csv',
                               index=False)
        else:
            self.output=self.panel()
            self.output.to_csv('Output/return2_'+str(self.n)+'_adv_days.csv',
                               index=False)
    
    def panel(self):
        self.comb=self.df[['AudienceType','AdDigitalChannel']].drop_duplicates()
        self.comb=self.comb.reset_index(drop=True)
        
        self.df_li=[]
        for i in range(len(self.comb)):
            new_df=self.df[(self.df['AudienceType']==self.comb['AudienceType'][i])&
                           (self.df['AdDigitalChannel']==self.comb['AdDigitalChannel'][i])]
            x=self.ndays_ad(new_df,self.n)
            self.df_li.append(x)
        return pd.concat(self.df_li,axis=0) 
        
    @staticmethod    
    def ndays(data,n):
        data=data.reset_index(drop=True)
        X=pd.DataFrame(columns=data.columns.tolist()[1:-1])
        Y=data['conversion'][n-1:].reset_index(drop=True)
        leng=len(data)
        for i in range(leng-n+1):
            temp_df=data.iloc[i:i+n,1:-2].sum()
            temp_df['CTR']=data.iloc[i:i+n,-2].mean()
            X=X.append(temp_df, ignore_index=True)
        return pd.concat([X,Y],axis=1)
    
    @staticmethod
    def ndays_ad(data,n):
        data=data.reset_index(drop=True)
        X=pd.DataFrame(columns=data.columns.tolist()[3:-1])
        Y=data['conversion'][n-1:].reset_index(drop=True)
        ID=data[['AudienceType','AdDigitalChannel']][n-1:].reset_index(drop=True)
        leng=len(data)
        for i in range(leng-n+1):
            temp_df=data.iloc[i:i+n,3:-2].sum()
            temp_df['CTR']=data.iloc[i:i+n,-2].mean()
            X=X.append(temp_df, ignore_index=True)
        return pd.concat([X,Y,ID],axis=1)        
```

```python
def ndays(data,n):
    data=data.reset_index(drop=True)
    X=pd.DataFrame(columns=data.columns.tolist()[3:-1])
    Y=data['conversion'][n-1:].reset_index(drop=True)
    leng=len(data)
    for i in range(leng-n+1):
        temp_df=data.iloc[i:i+n,3:-1].sum()
        X=X.append(temp_df, ignore_index=True)
    return pd.concat([X,Y],axis=1)
```

```python
def ndays_ad(data,n):
    data=data.reset_index(drop=True)
    X=pd.DataFrame(columns=data.columns.tolist()[3:-1])
    Y=data['conversion'][n-1:].reset_index(drop=True)
    ID=data[['CampaignID','AdDigitalChannel']][n-1:].reset_index(drop=True)
    leng=len(data)
    for i in range(leng-n+1):
        temp_df=data.iloc[i:i+n,3:-1].sum()
        X=X.append(temp_df, ignore_index=True)
    return pd.concat([X,Y,ID],axis=1)
```
